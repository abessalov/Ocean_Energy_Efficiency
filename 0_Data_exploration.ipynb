#### Other features

```python
feat = 'Part 1 Construction Age Band'
print(df[feat].isnull().sum())
df[feat].value_counts().sort_index().to_dict()
```

```python
# Construction age converting to integer
dict_repl = {
 'before 1919': 1919,
 '1919-1929': 1929,
 '1930-1949': 1949,
 '1950-1964': 1964,
 '1965-1975': 1975,
 '1976-1983': 1983,
 '1984-1991': 1991,
 '1992-1998': 1998,
 '1999-2002': 2002,
 '2003-2007': 2007,
 '2008 onwards': 2008,
}
feat = 'Part 1 Construction Age Band'
df[feat] = df[feat].map(dict_repl)
```

```python
# newest houses is more energy effective
_, ax = plt.subplots(1,2,figsize = (12,5))
_ = df.groupby(feat)['Primary Energy Indicator (kWh/m²/year)'].median().plot(kind = 'barh', ax = ax[0], ylabel = '')
_ = df.groupby(feat)['Primary Energy Indicator (kWh/m²/year)'].mean().plot(kind = 'barh', ax = ax[1], ylabel = '', color = 'green')
```

```python
# bigger house total area is more energy effective
df['x'] = pd.qcut(df['Total floor area (m²)'],q = 10)
_, ax = plt.subplots(1,2,figsize = (12,5))
_ = df.groupby('x')['Primary Energy Indicator (kWh/m²/year)'].median().plot(kind = 'barh', ax = ax[0], ylabel = '')
_ = df.groupby('x')['Primary Energy Indicator (kWh/m²/year)'].mean().plot(kind = 'barh', ax = ax[1], ylabel = '', color = 'green')
```
